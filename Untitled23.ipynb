```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import time 
```

```python
df = pd.read_csv(r"C:\depression\scores.csv")
```

```python
print(df)
```

```python
df["DeltaMadrs"] = df.madrs2 - df.madrs1
```

```python
df_conditions = df[df.number.str.contains('condition')].copy()
df_control = df[df.number.str.contains('control')].copy()
```

```python
df_conditions
```

```python
df_conditions.shape
```

```python
txt_missing = "_MISSING"
```

```python
#prep for melanch
df_conditions.melanch = df_conditions.melanch.fillna(txt_missing)
df_conditions.melanch = df_conditions.melanch.astype("category")
df_conditions.melanch = df_conditions.melanch.cat.rename_categories({-1 : txt_missing,
                                                                  1.0 : "1",
                                                                  2.0 : "2"})

df_conditions.age = df_conditions.melanch.astype("category")
df_conditions.gender = df_conditions.melanch.astype("category")


df_conditions.afftype = df_conditions.afftype.astype(int).astype("category")
df_conditions.inpatient = df_conditions.inpatient.astype(int).astype("category")
df_conditions.marriage = df_conditions.marriage.astype(int).astype("category")
df_conditions.work = df_conditions.work.astype(int).astype("category")

df_conditions.edu = df_conditions.edu.astype('category')
df_conditions.edu = df_conditions.edu.cat.rename_categories({' ' : txt_missing})
```

```python
df_conditions
```

```python
feature_num =['days','madrs1','madrs2','DeltaMadrs']
feature_cat =['age','marriage','work','melanch','afftype','inpatient','gender','edu']
```

```python
df_conditions[feature_num].describe()
```

```python
temp_plot_paras = plt.rcParams['figure.figsize']

plt.rcParams['figure.figsize'] = (14,4)
df_conditions.plot(x='number', y=['madrs1','madrs2'], kind='bar')
plt.title('MADRS Development')
plt.grid()
plt.show()

plt.rcParams['figure.figsize'] = temp_plot_paras
```

```python
for f in feature_num:
    df_conditions[f].plot(kind="hist")
    plt.title(f)
    plt.grid()
    plt.show()
    
```

```python
sns.pairplot(df_conditions[feature_num],kind = "reg")
plt.show()
```

```python
df_conditions[feature_num].corr(method='spearman')
```

```python
df_conditions[feature_num].corr(method='pearson')
```

```python
for f in feature_cat:
    plt.figure(figsize=(10,4))
    sns.violinplot(data=df_conditions, x=f, y='madrs1')
    plt.title('madrs1 vs ' + f)
    plt.grid()
    plt.show()
```

```python
for f in feature_cat:
    plt.figure(figsize=(10,4))
    sns.violinplot(data=df_conditions, x=f, y='madrs2')
    plt.title('madrs2 vs ' + f)
    plt.grid()
    plt.show()
```

```python
# impact of feature on score difference DeltaMADRS = madrs2 - madrs1
for f in feature_cat:
    plt.figure(figsize=(10,4))
    sns.violinplot(data=df_conditions, x=f, y='DeltaMadrs')
    plt.title('DeltaMADRS vs ' + f)
    plt.grid()
    plt.show()
```
